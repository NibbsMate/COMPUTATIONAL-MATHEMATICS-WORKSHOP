Plot the trigonometric function

```python
import matplotlib.pyplot as plt

import math


# Step 1: Generate equidistant x-coordinates between the given interval

# Lists for coordinates

x = []

y = []


# Define start, end, and step size for x increment

start = -4 * math.pi

end = 4 * math.pi

step = 0.1 # Controls smoothness of the graph


i = start

while i <= end:
    x.append(i) # Store x-coordinate

    # Step 2: Calculate the y-coordinate using sin function for each x

    y.append(math.sin(i))

    i += step


# Step 3: Plot the graph using calculated coordinates

# Label the graph properly

plt.plot(x, y, color="red", label="sin function")

plt.xlabel("X")

plt.ylabel("Y")

plt.title("y = sin(x) ,-4π<= x <= 4π")

plt.grid(True)

plt.legend()

plt.show()
```

# Circle

```python
x =[] # List to store x values

y= [] # List to store y values

r = 4

center = [2,2]

# paramater theta start value

theta_start = 0.0

# parameter Theta end value (bit more than 2pi)

theta_end = 2*math.pi

# step sizes to sample the paramater Theta

step_size= 0.125

#math.ceil() → round up to nearest integer

# Number of samples

steps = math.ceil((theta_end-theta_start)/step_size)

# initialize the theta

# variable to increment Theta by each step

theta = theta_start

# for loop to calculate x and y value at each steps

for i in range(0,int(steps)):
    x.append(r*math.cos(theta)+center[0])
    y.append(r*math.sin(theta)+center[1])
    theta += step_size

# create figure in new window

plt.figure()

# plot a dot a center

plt.plot(center[0],center[1],'.')

plt.plot(x,y,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

#plt.gca() → Get the current figure (axes object) for handling

ax =plt.gca()

#ensures 1 unit in x = 1 unit in y

#Without this ,figure is stretched(ircle could appear as an ellipse)

ax.set_aspect('equal',adjustable='box')

plt.show()
```

# Ellipse

```python
# importing a library

import matplotlib.pyplot as plt

import math

# Lists to store x and y values

x = []

y= []

# Ellipse parameters

a=3 # semi-minor axis

b=5 # semi-major axis

center =[2,2] # center (p,q)


# Parameter theta

theta_start = 0.0

theta_end = 3*math.pi # make more than 2*pi

step_size = 0.125


# Number of samples

steps = math.ceil((theta_end -theta_start) / step_size)


# initializing theta

theta= theta_start # calculate x and y cooradinates

for i in range(int(steps)):
    x.append(a*math.cos(theta)+center[0]) # x= a*cos(θ) +p
    y.append(b*math.sin(theta)+center[1]) # y=b*sin(θ) +q
    theta += step_size # incremet the theta by step_size


# Plotting

plt.figure()

# plot the center point

plt.plot(center[0],center[1],'.')

#Plot the ellipse

plt.plot(x,y,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

plt.title("Ellipse centered at (2,2) with a=3 , b=5")


# Equal aspect ration to avoid stretching

ax= plt.gca()

ax.set_aspect('equal',adjustable='box')

plt.show()
```

# Parabola:

```python
import matplotlib.pyplot as plt

import math

# parabola parameters

q=0 # y-coordinate of vertex

p=0 # x-coordinate of vertex

a=0.5 # controls width and direction

# Step size for x values

step_size = 0.1

# define x range aroung vertex

y_start = p-5

y_end = p + 5

# lists to store coordinates

x_values =[]

y_values =[]

#initialize x

y = y_start

# generate poinst for parabola while y <= y_end:

x = 4*a*(y-p)**2+p

x_values.append(x)

y_values.append(y)

y += step_size

#plot the graph

plt.figure()

plt.title(f"Parabola with vertex at ({p},{q}) and a ={a}")

plt.plot(x_values,y_values,'-')

plt.grid()

plt.xlabel("X")

plt.ylabel("Y")

plt.show()
```

# Circle

10. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop.

```python
import matplotlib.pyplot as plt
import math

# center and radius
a = 2
b = 2
r = 4

x = []
y = []

theta = 0

# generate points using while loop
while theta <= 2 * math.pi:
    x.append(a + r * math.cos(theta))
    y.append(b + r * math.sin(theta))
    theta += 0.1   # step size

# plot circle
plt.plot(x, y)

plt.gca().set_aspect('equal')  # makes it a perfect circle
plt.title("Circle with center (2,2) and radius 4")

plt.show()
```

# Ellipse

11. Plot the above circle and the ellipse on the same axes. This is give an idea about the intersection points of the two conic sections.

```python
import matplotlib.pyplot as plt
import math

# Circle parameters
a = 2
b = 2
r = 4

# Ellipse parameters (example)
rx = 6   # horizontal radius
ry = 3   # vertical radius

# lists
x_circle = []
y_circle = []

x_ellipse = []
y_ellipse = []

theta = 0

# generate points
while theta <= 2 * math.pi:
    # circle
    x_circle.append(a + r * math.cos(theta))
    y_circle.append(b + r * math.sin(theta))

    # ellipse
    x_ellipse.append(a + rx * math.cos(theta))
    y_ellipse.append(b + ry * math.sin(theta))

    theta += 0.1

# plot both
plt.plot(x_circle, y_circle, label="Circle")
plt.plot(x_ellipse, y_ellipse, label="Ellipse")

plt.gca().set_aspect('equal')
plt.legend()
plt.title("Circle and Ellipse on Same Axes")

plt.show()
```
