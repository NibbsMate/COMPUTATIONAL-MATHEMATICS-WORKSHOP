Task 4a: np.dot() vs. np.matmul()
The primary difference between np.dot() and np.matmul() lies in how they handle higher-dimensional arrays (tensors).

Handling of Scalars: np.dot() can handle scalar multiplication, while np.matmul() does not allow scalars and will throw an error.

Higher Dimensions: For arrays with more than 2 dimensions, np.dot() performs a specific sum-product over the last axes, whereas np.matmul() treats the arrays as "stacks" of matrices and performs a broadcasted matrix multiplication.

Result of np.dot(5, arr1): If arr1 is [1, 2, 3], the result is [5, 10, 15]. It acts as a scalar multiplication, multiplying every element in the array by 5.

## Task 6a: 3D Vector Plotting
Create three vectors in 3D space and plot them using Matplotlib.

```python
# Define vectors
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

print("Vectors:")
print(f"U = {U}")
print(f"V = {V}")
print(f"W = {W}")
```

```python
# Create 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

origin = [0, 0, 0]

# Plot vectors
ax.quiver(origin[0], origin[1], origin[2], U[0], U[1], U[2], color='r', arrow_length_ratio=0.1, linewidth=2, label='U')
ax.quiver(origin[0], origin[1], origin[2], V[0], V[1], V[2], color='g', arrow_length_ratio=0.1, linewidth=2, label='V')
ax.quiver(origin[0], origin[1], origin[2], W[0], W[1], W[2], color='b', arrow_length_ratio=0.1, linewidth=2, label='W')

# Set labels and limits
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([0, 1.5])
ax.set_ylim([0, 1.5])
ax.set_zlim([0, 1.5])
ax.set_title('3D Vectors', fontsize=14)
ax.legend()

plt.tight_layout()
plt.show()

print("✓ 3D vectors plotted successfully")
```

## Task 6b: 2D Vector Addition
Create two vectors in 2D space, add them to create a third vector, and plot all three vectors.

```python
# Define 2D vectors
U_2d = np.array([-0.5, 0.5])
V_2d = np.array([1, 1])

# Add vectors
W_2d = U_2d + V_2d

print(f"U = {U_2d}")
print(f"V = {V_2d}")
print(f"W = U + V = {W_2d}")
```

```python
# Create 2D plot
fig, ax = plt.subplots(figsize=(8, 8))
origin = [0, 0]

# Plot vectors
ax.quiver(origin[0], origin[1], U_2d[0], U_2d[1], angles='xy', scale_units='xy', scale=1, color='r', width=0.01, label='U')
ax.quiver(origin[0], origin[1], V_2d[0], V_2d[1], angles='xy', scale_units='xy', scale=1, color='g', width=0.01, label='V')
ax.quiver(origin[0], origin[1], W_2d[0], W_2d[1], angles='xy', scale_units='xy', scale=1, color='b', width=0.01, label='W=U+V')

# Add labels
ax.text(-0.3, 0.5, 'U', fontsize=12, color='red', fontweight='bold')
ax.text(1.1, 1.1, 'V', fontsize=12, color='green', fontweight='bold')
ax.text(0.6, 1.6, 'W', fontsize=12, color='blue', fontweight='bold')

# Set limits and labels
ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2.5])
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_title('2D Vector Addition: W = U + V', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()

print("✓ 2D vectors plotted successfully")
```

## Task 6c: Vector Magnitude
Write a function to calculate magnitude of a vector.

```python
def magnitude(v):
    """
    Calculate the magnitude (length) of a vector
    Magnitude = √(x² + y² + z²)
    """
    return np.sqrt(np.sum(v**2))

# Calculate magnitude
U_mag = np.array([1.5, 1.5, 2.0])
mag = magnitude(U_mag)

print(f"Vector U = {U_mag}")
print(f"\nMagnitude Calculation:")
print(f"||U|| = √(1.5² + 1.5² + 2.0²)")
print(f"||U|| = √(2.25 + 2.25 + 4.0)")
print(f"||U|| = √8.5")
print(f"||U|| = {mag:.4f}")
```

## Task 6d: Direction Cosines
Write a function to calculate the direction cosines of a vector.

```python
def direction_cosines(v):
    """
    Calculate direction cosines of a vector
    Direction Cosines = v / magnitude(v)
    """
    return v / magnitude(v)

# Calculate direction cosines
O = np.array([1.5, 1.5, 2.0])
mag = magnitude(O)
cos_dir = direction_cosines(O)

print(f"Vector O = {O}")
print(f"\nMagnitude = {mag:.4f}")
print(f"\nDirection Cosines:")
print(f"cos(α) = {O[0]}/{mag:.4f} = {cos_dir[0]:.4f}")
print(f"cos(β) = {O[1]}/{mag:.4f} = {cos_dir[1]:.4f}")
print(f"cos(γ) = {O[2]}/{mag:.4f} = {cos_dir[2]:.4f}")
print(f"\nDirection Cosines = {cos_dir}")
```

## Task 6e: Scalar (Dot) Product
Write a function to calculate the scalar (dot) product of two vectors.

```python
def dot_product(a, b):
    """
    Calculate dot product of two vectors
    a · b = a₁b₁ + a₂b₂ + a₃b₃
    """
    return np.dot(a, b)

# Calculate dot product
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
dot = dot_product(a, b)

print(f"Vector a = {a}")
print(f"Vector b = {b}")
print(f"\nDot Product Calculation:")
print(f"a · b = (1.5)(-1.5) + (1.5)(0.5) + (2.0)(2.0)")
print(f"a · b = -2.25 + 0.75 + 4.0")
print(f"a · b = {dot:.2f}")
```

## Task 6f: Vector (Cross) Product
Write a function to calculate the vector (cross) product of two vectors.

```python
def cross_product(a, b):
    """
    Calculate cross product of two vectors
    a × b = |i    j    k   |
             |a₁   a₂   a₃  |
             |b₁   b₂   b₃  |
    """
    return np.cross(a, b)

# Calculate cross product
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
cross = cross_product(a, b)

print(f"Vector a = {a}")
print(f"Vector b = {b}")
print(f"\nCross Product Calculation:")
print(f"a × b = |i     j     k   |")
print(f"        |1.5   1.5   2.0 |")
print(f"        |-1.5  0.5   2.0 |")
print(f"\na × b = [i(1.5×2.0 - 2.0×0.5) - j(1.5×2.0 - 2.0×(-1.5)) + k(1.5×0.5 - 1.5×(-1.5))]")
print(f"a × b = [i(3.0 - 1.0) - j(3.0 + 3.0) + k(0.75 + 2.25)]")
print(f"a × b = [i(2.0) - j(6.0) + k(3.0)]")
print(f"\na × b = {cross}")
```
