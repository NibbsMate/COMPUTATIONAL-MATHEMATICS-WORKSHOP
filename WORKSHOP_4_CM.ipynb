**TASK 2**
Following data shows the change in population of different animals from 2017 to 2022.
a. Create a pie chart for the population of each year
b. Can we show all the pie chart on the single pane? Write the code.a.png

```python
import matplotlib.pyplot as plt

# 1. Prepare the data
years = [2017, 2018, 2019, 2020, 2021, 2022]
animals = ['Bears', 'Dolphins', 'Whales']

# Data organized by year (rows from your table)
data = [
    [8, 150, 80],   # 2017
    [54, 77, 54],   # 2018
    [93, 32, 100],  # 2019
    [116, 11, 76],  # 2020
    [137, 6, 93],   # 2021
    [184, 1, 72]    # 2022
]

# 2. Create a figure with a grid of 2 rows and 3 columns
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.flatten()  # Flatten to easily loop through them

# 3. Loop through data to create each pie chart
for i in range(len(years)):
    axes[i].pie(data[i], labels=animals, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'Population in {years[i]}')

# Adjust layout so charts don't overlap
plt.tight_layout()
plt.show()
```

**Task 3:
Construct a pie chart to visually display the favorite fruits of the students in a class based on the
given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30**

```python
import matplotlib.pyplot as plt

# 1. Prepare the data
fruits = ['Mango', 'Orange', 'Plum', 'Pineapple', 'Melon']
counts = [45, 30, 15, 30, 30]

# 2. Create the pie chart
plt.pie(counts, labels=fruits, autopct='%1.1f%%', startangle=140)

# 3. Add a title
plt.title('Favorite Fruits of Students')

# 4. Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# The Data
x = ["Red", "Silver", "Blue", "White", "Green"]
y = [123, 230, 78, 193, 12]

# a) Basic Bar Chart
plt.bar(x, y)
plt.title("Car Colors Seen")
plt.show()

# b) Bar Chart with specific width
# Effect: The bars become narrower (0.3 of the default width).
plt.bar(x, y, width=0.3)
plt.title("Narrow Bar Chart")
plt.show()

# c) Horizontal Bar Chart
# Effect: The axes swap; categories are on the y-axis and values on the x-axis.
plt.barh(x, y)
plt.title("Horizontal Bar Chart")
plt.show()

# d) Bar Chart with matching colors
bar_colors = ["red", "silver", "blue", "white", "green"]
# We add an edge color for 'white' so it is visible against a light background
plt.bar(x, y, color=bar_colors, edgecolor="black")
plt.title("Colored Bar Chart")
plt.show()

# e) Pie Chart with Title
plt.pie(y, labels=x)
plt.title("Distribution of Car Colors")
plt.show()

# f) Testing additional Pie Chart arguments
# Note: 'myexplode' must match the length of the data (5 items)
myexplode = [0.3, 0, 0, 0, 0]
mycolours = ["Magenta", "Green", "Cyan", "Red", "Gray"]

# Example combining A, B, and D:
plt.pie(y, labels=x, explode=myexplode, shadow=True, colors=mycolours)
plt.legend(title="Car Colours:") # Requirement C
plt.show()
```
