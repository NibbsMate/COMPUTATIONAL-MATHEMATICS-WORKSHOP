a. Write a function fib (n) that prompts the number of terms “n” from the user and
displays first “n” terms of Fibonacci sequence iteratively.

```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        # Update a to be b, and b to be the sum of both
        a, b = b, a + b

# Prompt the user for input
num_terms = int(input("Enter the number of terms: "))
fib(num_terms)
```

b. Write a function myPow(x, n) that returns 𝑥𝑛, where and n is a non-negative integer.
Do not use the ** operator or the math.pow function - use one while loop.

```python
def myPow(x, n):
    result = 1
    count = 0

    # Loop until we have multiplied x exactly n times
    while count < n:
        result = result * x
        count = count + 1

    return result

# Example call:
print(myPow(2, 3))  # This will output 8 (2 * 2 * 2)
```

c. Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

```python
def SumDigits(n):
    total = 0
    # Convert number to string to loop through each digit
    for digit in str(n):
        total += int(digit)

    print(f"The sum of digits is: {total}")

# Testing the function
SumDigits(234)
```

d. Write a function squaredSum that returns the sum of the squares of an array/list of
integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python
to show the logic in your solution, do not use any library functions.


```python
def squaredSum(numbers):
    total = 0

    # Loop through every number in the list
    for num in numbers:
        # Square the number and add it to the total
        total = total + (num * num)

    return total

# Example usage:
result = squaredSum([2, 3, 4])
print(result)  # Output: 29
```

e. Write a function called sumDigits that is given an integer num and returns the sum of
the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if
there is one).


```python
def sumDigits(num):
    total = 0
    for i in range(1, num + 1):
        total = total + i
    return total

print(sumDigits(6)) # Output: 21
```
